```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv("/content/drive/MyDrive/FYP/CAA_NRC/CAA_NRC-preprocessed.csv")
df.head()
```

```python
text=df["new_text"]
coun_vect = CountVectorizer(max_features = 10000,stop_words='english')
count_matrix = coun_vect.fit_transform(text)
```

```python
count_array = count_matrix.toarray()
count_array
```

```python
df=pd.concat([df, pd.DataFrame(count_array)], axis=1)
```

```python
df.to_csv("/content/drive/MyDrive/FYP/CAA_NRC/BagOfWords/CAA_NRC-countvectors.csv",index=False)
```
