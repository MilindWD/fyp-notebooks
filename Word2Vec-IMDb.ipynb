```python
import pandas as pd
url="/content/gdrive/MyDrive/FYP/IMDb/IMDB-preprocessed.csv"
df=pd.read_csv(url)
df.head()
```

```python
text=df["new_text"]
sentences=[]
for i in text:
  i=i.split(" ") 
  sentences.append(i)
```

```python
sentences[0]
import gensim
w2v_model=gensim.models.Word2Vec(sentences=sentences,size=300,window=10,min_count=1)
```

```python
w2v_model.train(sentences,epochs=10,total_examples=len(sentences))
```

```python
w2v_model.save("/content/gdrive/MyDrive/FYP/IMDb/Word2Vec/w2c")
```
