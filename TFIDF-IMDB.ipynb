```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/FYP/IMDb/IMDB-preprocessed.csv")
df.head()
```

```python
X=df["new_text"]
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, norm='l2', encoding='latin-1', ngram_range=(1, 2),max_features = 5000,stop_words='english')
import pickle
features = tfidf.fit(X)
pickle.dump(features, open("tfidf.pickle", "wb"))
features=features.transform(X)
print(features)

```

```python
features = features.toarray()
df=pd.concat([df, pd.DataFrame(features)], axis=1)
```
