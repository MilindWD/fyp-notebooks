```python
import pandas as pd
df=pd.read_csv("/content/gdrive/MyDrive/FYP/IMDb/IMDB-preprocessed.csv")
df.head()
```

```python
text=df["new_text"]
sentences=[]
for i in text:
  i=i.split(" ") 
  sentences.append(i)
```

```python
from gensim.models.fasttext import FastText
```

```python
embedding_size = 300
window_size = 10
```

```python
ft_model = FastText(sentences,
                      size=embedding_size,
                      window=window_size,
                      iter=10)
```

```python
ft_model.save("/content/gdrive/MyDrive/FYP/IMDb/FastText/ft")
```
