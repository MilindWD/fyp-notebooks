```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
def display_metrics(y_test,y_pred):
    print(accuracy_score(y_test, y_pred))
    print(f1_score(y_test, y_pred, average="macro"))
    print(precision_score(y_test, y_pred, average="macro"))
    print(recall_score(y_test, y_pred, average="macro"))
```

```python
import pandas as pd
url="/content/gdrive/MyDrive/FYP/CAA_NRC/Tf-Idf/CAA_NRC-tfidf.csv"
df=pd.read_csv(url)
```

```python
y=df["label"]
X=df.iloc[:,2:]
X.shape
```

```python
import numpy as np
X=np.array(X)
X=X.reshape(3101, 5000,1)
X=X.astype('float32')
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Bidirectional,LSTM,GRU
# define the model
model = Sequential()
model.add(LSTM(10,activation="sigmoid",dtype='float32'))
model.add(Dense(128,activation="sigmoid"))
#model.add(Dense(64,activation="sigmoid"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=256, verbose=1)
```

```python
y_pred=model.predict(X_test)
```

```python
for i in range(len(y_pred)):
  if y_pred[i]>0.5:
    y_pred[i]=1
  else:
    y_pred[i]=0
display_metrics(y_test,y_pred)
```
