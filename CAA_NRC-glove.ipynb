```python
from glove import Corpus, Glove
```

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/FYP/CAA_NRC/CAA_NRC-preprocessed.csv")
df.head()
```

```python
text=df["new_text"]
sentences=[]
for i in text:
  i=i.split(" ") 
  sentences.append(i)
```

```python
word_corpus = Corpus() 

word_corpus.fit(sentences, window=10)
glove = Glove(no_components=300)
 
glove.fit(word_corpus.matrix, epochs=10, no_threads=4, verbose=True)
glove.add_dictionary(word_corpus.dictionary)
glove.save('/content/drive/MyDrive/FYP/CAA_NRC/Glove/glove.model')
```
