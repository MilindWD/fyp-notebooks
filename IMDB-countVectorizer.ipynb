```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/Fyp/IMDB-preprocessed.csv", names=['sentiment', 'new_text'], encoding='latin-1')
# df = df[1:]
df.head()
```

```python
text=df["new_text"]
coun_vect = CountVectorizer(max_features = 3000,stop_words='english')
count_matrix = coun_vect.fit_transform(text)
```

```python
count_array = count_matrix.toarray()
count_array
```

```python
count_array.shape
```

```python
df=pd.concat([df, pd.DataFrame(count_array)], axis=1)
df = df[1:]
df
```

```python
df.sentiment.value_counts()
```

```python
df.to_csv("/content/gdrive/MyDrive/Fyp/CountVectorizer/IMDB_countVectorsFull.csv",index=False)
```
