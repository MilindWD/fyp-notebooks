```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
def display_metrics(y_test,y_pred):
    print(accuracy_score(y_test, y_pred))
    print(f1_score(y_test, y_pred, average="macro"))
    print(precision_score(y_test, y_pred, average="macro"))
    print(recall_score(y_test, y_pred, average="macro"))
```

```python
import pandas as pd
url="/content/gdrive/MyDrive/FYP/IMDb/Tf-Idf/IMDB-tfidf.csv"
df=pd.read_csv(url)
```

```python
y=df["sentiment"]
X=df.iloc[:,2:]
X.shape
```

```python
import numpy as np
X=np.array(X)
X=X.reshape(50000, 5000,1)
X=X.astype('float32')
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Bidirectional,LSTM,GRU,Dropout
# define the model
model = Sequential()
model.add(LSTM(128,activation="relu",dtype='float32',return_sequences=True))
model.add(LSTM(128,activation="relu",dtype='float32',return_sequences=False))
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.60))
model.add(Dense(16,activation='relu'))
model.add(Dropout(0.60))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, batch_size=64, verbose=1)
model.save('/content/gdrive/MyDrive/LSTM_TFIDF_model')


```

```python
y_pred=model.predict(X_test)
```

```python
for i in range(len(y_pred)):
  if y_pred[i]>0.5:
    y_pred[i]=1
  else:
    y_pred[i]=0
display_metrics(y_test,y_pred)
```
